```python
# imports
import numpy as np
import scipy.constants as pc
import matplotlib.pyplot as plt
import matplotlib.colors as c
import sympy as sy
from sympy import *
```

```python
fig, ax = plt.subplots(figsize=(10,4))
std_y_g =  np.array([89.5])
std_y_w = np.array([329.87])

x = np.array([1, 2, 3, 4])
y_w = np.array([1001, 1001, 1001, 1001])
y_g = np.array([1480, 1480, 1480, 1480])

ax.errorbar(2, 925.00, xerr=0, yerr=std_y_g, fmt='D', color='b', label=r'Errorleiste und Bestwert $ \rho_G $ ')
ax.errorbar(3, 919.44, xerr=0, yerr=std_y_w, fmt='D', color='r', label=r'Errorleiste und Bestwert $ \rho_W $ ')
ax.plot(x,y_g,color='b',label=r'Literaturwert $ \rho_G $ ')
ax.plot(x,y_w,color='r',label=r'Literaturwert $ \rho_W $ ')
ax.set(xlabel='Viskosität', ylabel='Unbedeutend/Einheitenlos', title='Gemessene Viskositäten und Literaturwerte')
ax.legend()
ax.grid()
fig.savefig("vergleich lit.png")
```
